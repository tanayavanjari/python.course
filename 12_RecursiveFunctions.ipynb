# **Recursive Functions**

A recursive function is a function that calls itself to solve a problem by breaking it into smaller sub-problems.

**Key Parts of a Recursive Function**

Every recursive function must have two important parts:

1. Base Case

The condition where the function stops calling itself

Prevents infinite recursion

2. Recursive Case

The part where the function calls itself

Moves the problem closer to the base case

**Factorial of a number**

5! = 5 X 4 X 3 X 2 X 1

5! = 120

n! = n X (n-1)!
  
= n X (n-1)! X (n-2)!

= so on... till it reaches to 1

1! = 1 --> Base Case

**factorial(3)**

→ 3 × factorial(2)

→ 3 × 2 × factorial(1)

→ 3 × 2 × 1

→ 6

# **Factorial**
**WAP to calculate factorial of a number using user defined function.**

Mathematically 0! = 1, 1! = 1

```python
def factorial(n):
  if n == 0 or n == 1: # base case
    return 1
  else:
    return n * factorial(n-1) # Recursive case
```

```python
factorial(3)
```

```python
factorial(5)
```

# **Sum of first n numbers using Recursion**

eg. If I pass n=4 it should find 4+3+2+1=10, It should stop when n=0

```python
def sum_natural_nos(n):
  if n==0:
    return 0
  else:
    return n + sum_natural_nos(n-1)
```

```python
sum_natural_nos(0)
```

```python
sum_natural_nos(4) # 4 + 3 + 2 + 1 = 10
```

```python
sum_natural_nos(5) # 5 + 4 + 3 + 2 + 1 =15
```

# **Fibonacci Series**

The Fibonacci series is a sequence where each number is the sum of the previous two numbers.

WAP to print Fibonacci Series using recursion.

fib(0) = 0        ← base case

fib(1) = 1        ← base case

fib(n) = fib(n-1) + fib(n-2)   ← recursive case

```python
# Write a function which returns the nᵗʰ Fibonacci number using recursion.
def fibonacci(n):
  if n<= 0:
    return "Invalid Input"
  elif n == 1:
    return 0 # 1st fibonacci number is 0
  elif n == 2:
    return 1 # 2nd fibonacci number is 1
  else:
    return fibonacci(n-1) + fibonacci(n-2) # Recursive Call
```

**How above program works?**

fibonacci(6)

= fibonacci(5) + fibonacci(4)

= [(fibonacci(4) + fibonacci(3))] + [(fibonacci(3) + fibonacci(2))]

= [(fibonacci(3) + fibonacci(2))] + [(fibonacci(2) + fibonacci(1))] + [(fibonacci(2) + fibonacci(1))] + 1

= [(fibonacci(2) + fibonacci(1))] + 1 + 1 + 0 + 1 + 0 + 1

= 1 + 0 + 1 + 1 + 0 + 1 + 0 + 1

= 5

```python
# fibonacci(4)
fibonacci(4)
```

```python
# fibonacci(5) = 0 1 1 2 3
fibonacci(5)
```

```python
# fibonacci(6) = 0 1 1 2 3 5
fibonacci(6)
```

```python
# fibonacci(7) = 0 1 1 2 3 5 8
fibonacci(7)
```

**Print first n Fibonacci numbers**

```python
n = int(input("Enter number of terms: "))

for i in range(1,n+1): # each value of i represents the position in the fibonacci sequence
    print(fibonacci(i), end=" ")
```

# **String Reversal using Recursion**



# **How String Reversal Works**

reverse_string("abc")

**Execution Flow:**

("abc"[1:]) + "abc"[0])

reverse_string("abc")
→ reverse_string("bc") + "a"

reverse_string("bc")
→ reverse_string("c") + "b" + "a"

reverse_string("c")
→ reverse_string("") + "c" + "b" + "a"

reverse_string("")
→ ""

**Now return back:**

"" + "c" = "c"

"c" + "b" = "cb"

"cb" + "a" = "cba"

**Final Output:**

"cba"

```python
def reverse_string(s):
  if len(s) == 0: # Base Case - This stops the recursion.
                  # When the string is empty, return it as it is.
    return s
  else:
    return reverse_string(s[1:]) + s[0] # "hello"[1:] → "ello"
```

```python
print(reverse_string("Hello"))
```

```python
print(reverse_string("ab"))
```

```python
print(reverse_string("Neha"))
```
