# **Dictionary**

*    Dictionaries are used to store data values in "key:value pairs".
*   Dictionaries are written with curly brackets {"Key":"Value"}, and have keys and values.
*   A dictionary is a collection which is ordered, changeable and do not allow duplicates.

**Additional Functions**

```python
# get(): If Key exists → returns its value
student = {"name": "Neha", "age": 21}
print(student.get("name"))
```

```python
# If Key does NOT exist → returns None. No Error raised
print(student.get("marks"))
```

```python
student["marks"]
```

```python
# Using a default value (very important)
# If the key is not found, it returns None or a specified default value.
print(student.get("marks", 0))
```

```python
# Difference between get() and []
# To extract value from dictionary we use key name in [].
# But in this case error will be raised if key is not present
print(student["marks"])
```

```python
# using get() we are safe as no error will be raised
print(student.get("marks"))
```

```python
# Copy(): creates a shallow copy of dictionary
d1 = {"a": 1, "b": 2}
d2 = d1.copy()

print(d1)
print(d2)
```

```python
# Changes in d2 do not affect d1.
d2["a"] = 100
print(d1)
print(d2)
```

```python
# fromkeys(): Creates a new dictionary using given keys and a common value.
# default value is None
keys = ("name", "age", "marks") # only keys of dictionary, no values
student = dict.fromkeys(keys)

print(student)
```

```python
# with a default value
keys = ("a", "b", "c")
d = dict.fromkeys(keys, 0)

print(d)
```

```python
# setdefault(): Returns the value of a key.
# If the key does not exist, it adds the key with a default value.
student = {"name": "Snehal", "age": 21}

print(student.setdefault("name", "Unknown"))
print(student)
```

```python
# if key does not exist it adds key with a default value 80
print(student.setdefault("marks", 80))
print(student)
```

```python
# get(): Returns default, does not add key
# setdefault(): returns value of key; inserts key if missing
# fromkeys(): creates a new dictionary from given keys
# copy(): creates a shallow copy of a dictionary
```

## **Sets**
*   A set is an unordered collection of unique elements.
*   i.e. each element in a set must be unique.
*   Sets are mutable, meaning you can add or remove items after a set has been created.
*   Sets are defined using curly braces {} or the built-in set() function.





```python
# Create a Set
set1 = {1, 2, 3, 4, 5}
print (set1)
```

```python
# Create set using built-in set() function
set2 = set([1, 2, 3, 4, 5])
print (set2)
```

```python
set2 = set((1, 2, 3, 4, 5))
print (set2)
```

```python
set3 = {1, 2, 2, 3, 3, 4, 5, 5}  # duplicates will automatically be removed
print (set3)
```

### **Set Operations**
*   Union − It combine elements from both sets using the union() function or the | operator.
*   Intersection − It is used to get common elements using the intersection() function or the & operator.
*   Difference − It is used to get elements that are in one set but not the other using the difference() function or the - operator.
*   Symmetric Difference − It is used to get elements that are in either of the sets but not in both using the symmetric_difference() method or the ^ operator.





**Union: It will merge both sets and duplicate elemets will be entered once**

```python
set1={"English","Hindi","Marathi"}
set2={"Tamil","French","English"}
new_set = set1.union(set2)
print(new_set)
```

```python
var1= set1 | set2
print(var1)
```

**Intersection: Common elements in both sets**

```python
set1={"English","Hindi","Marathi"}
set2={"Tamil","French","English"}
data=set1.intersection(set2)
print(data)
```

```python
data2=set1 & set2
print(data2)
```

```python
set1
```

```python
set2
```

**intersection_update() - Store the result of intersection() in set1**

```python
set1.intersection_update(set2)
print(set1)
print(set2)
```

**Symmetric_difference() : it is a opposite of intersection()**
*   Except common elements put all elements in new set



```python
set1={'a','b','c'}
set2={'a','d','e'}
new_set=set1.symmetric_difference(set2)
print(new_set)
```

```python
new_set2=set1 ^ set2
print(new_set2)
```

**In set2 save the resultant of symmetric_difference() using symmetric_difference_update()**

```python
set2.symmetric_difference_update(set1)
print(set2)
print(set1)
```

**Frozen Set: A frozen set is an immutable collection of unique elements, similar to a regular set but with the distinction that it cannot be modified after creation. Once created, the elements within a frozen set cannot be added, removed, or modified, making it a suitable choice when you need an immutable set.**

```python
set_a=frozenset((12,34,56,78,90,2,3,4))
print(set_a)
```
