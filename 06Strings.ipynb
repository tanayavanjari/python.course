```python
# Each character of a word has an index
s = "Python"
print(s[0])   # P
print(s[-1])  # n
print(s[5])
```

```python
# find(): find() is a string method used to search for a substring inside a string.
fruit = "banana"
fruit.find('a') # 1st occurence
```

```python
fruit.find('a',2) # start index
```

```python
fruit.find('a',2,5) # start and end index
```

```python
# Substring not found â†’ returns -1
fruit.find('s')
```

```python
# islower()
s1 = "python"
s2 = "Python"
s3 = "123abc"
s4 = "123"

print(s1.islower())  # True
print(s2.islower())  # False
print(s3.islower())  # True
print(s4.islower())  # False
```

```python
# isupper()
s1 = "PYTHON"
s2 = "Python"
s3 = "ABC123"
s4 = "123"

print(s1.isupper())  # True
print(s2.isupper())  # False
print(s3.isupper())  # True
print(s4.isupper())  # False
```

```python
# isspace():
x=" "
print(x.isspace())
```

```python
# startswith()
x="python"
print(x.startswith("p"))
```

```python
print(x.startswith("y"))
```

```python
# endswith():
print(x.endswith("n"))
```

```python
print(x.endswith("z"))
```

```python
# join(): Joins elements of an iterable of strings using a separator string.
s1 = "Snehal"
s2 = "-"      # "_" is a separator
s2.join(s1)
```

```python
words = ["I", "love", "Python"]
result = " ".join(words)    # " " is a separator
print(result)
```

```python
# strip(): Removes leading and trailing spaces (or specified characters).
name = "    Snehal    "
print(name.strip())
```

```python
# lstrip(): Removes leading (left-side) spaces only.
print(name.lstrip())
```

```python
# rstrip(): Removes trailing (right-side) spaces only.
print(name.rstrip())
```
