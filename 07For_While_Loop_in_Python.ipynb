# **For loop in Python**
# A for loop in Python is used to iterate (loop) over a sequence such as a list, tuple, string, set, dictionary, or range of numbers. It allows you to execute a block of code repeatedly for each element in the sequence.

Syntax:
for variable in sequence:
    statement(s)

variable → takes the value of each item in the sequence one by one
sequence → any iterable object
The loop ends automatically when the sequence is exhausted

*   For loop is used to execute selected statements multiple times.
*   If a statement is executed once, it is called an iteration.

# **While Loop in Python**
*   A while loop in Python is used to repeat a block of code as long as a given condition is True.
*   The condition is checked before each iteration.

If the condition is True, the loop body executes.

If the condition becomes False, the loop stops.

```python
# While loop: Print Hello for 5 times
cnt = 0
while cnt < 3:
  print("Hello")
  cnt = cnt + 1
```

```python
# Print all the numbers from 1 to 5
i = 1
while i <=5 : # OR i<6
    print(i)
    #i = i + 1 # OR i+=1
    i += 1
```

```python
# Print sum of first 5 numbers
i = 1
total = 0

while i <= 5:
    total = total + i
    i = i + 1

print("Sum =", total)
```

```python
# Print sum of numbers from 6 to 10
i = 6
total = 0

while i<=10:
  total = total + i
  i = i + 1

print("Sum =", total)
```

**While Loop with else**

The else block runs when the loop ends normally (not using break).

```python
i = 1
while i <= 3:
    print(i)
    i += 1
else:
    print("Loop completed")
```

**When to Use While Loop**

When the number of iterations is not known in advance

When looping depends on a condition

Example: menu-driven programs, user input validation

**continue Keyword in Python**

The continue keyword is used inside loops (for or while) to skip the current iteration and move to the next iteration of the loop.

```python
# Here is the same code rewritten using continue instead of break
# (so the loop does not stop, it only skips when i == 3)
for i in range(10):
    if i == 3:
        continue
    else:
        print(i)
```

**Using continue in while loop**

continue skips the current iteration.

```python
i = 0
while i < 5:
    i += 1
    if i == 3:
        continue
    print(i)
```

**pass Keyword in Python**

The pass keyword in Python is a null statement.
It does nothing when executed, but it is used where a statement is syntactically required.

```python
# pass
for i in range(5):
  if i==3:
    pass
  else:
    print(i)
```

### **Difference between pass and continue**

pass is a null statement used as a placeholder

continue skips the current iteration of a loop

continue can be used only inside loops

pass can be used anywhere a statement is required

```python
# pass does nothing, so print(i) still executes.
for i in range(1, 5):
    if i == 3:
        pass
    print(i)
```

```python
# When i == 3, print(i) is skipped.
for i in range(1, 5):
    if i == 3:
        continue
    print(i)
```

**Using break in Python**

break stops the loop immediately.

```python
# break
for i in range(10):
  if i==3:
    break
  else:
    print(i)
```

```python
i = 1
while i <= 10:
    if i == 5:
        break
    print(i)
    i += 1
```
