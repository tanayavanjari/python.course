## **Some of terminalogies / keywords used in functions**

- **`def`** - Marks the start of the function header.
- **`function name`** - used to uniquely identify the function. Function naming follows the same check list which we followed for variables.
- **`Params/Args`** - used to pass values to a function. Params are optional.
- **`Colon (:)`**- marks the end of the function header.
- **`Doc String`** - A short description about the function.This is optional.
- **`Business logic/Statements`** - One ore more valid Python statements to perform the required task.
- **`return`** - This is optional. But this statement will help you to return a value from the function.
- **`print`** - To dispaly the value from the function. This is optional.
Either print or return need to be included at the end of the function.
Make sure proper indentation is given inside the function body.

# **Compute the Hypotenuse**

Read the lengths of the two shorter sides of a right triangle as its parameters. Return the hypotenuse of the triangle, computed using Pythagorean theorem as the function's result.

c square = a square + b square

```python
def hypo(a,b):
  from math import sqrt
  a_square = a * a
  b_square = b * b
  hypotenuse = sqrt(a_square + b_square)
  return hypotenuse
```

```python
s1 = float(input("Enter length of side 1: "))
s2 = float(input("Enter length of side 2: "))
res = hypo(s1,s2)
print("%0.2f" %res)
```

# **Taxi Fare**
Taxi's base fare is dollar 4.00 plus dollar 0.25 for every 140 meters traveled.

Write a function that takes the distance traveled (in km) as its only parameter and returns the total fare as its only result.

WAP that demonstrates the function.


```python
# taxi fare
def taxi(D): # D in km
  d=D*1000 # convert km to meters
  price=4+(d/140)*0.25 # d/140 -> how many times 140 meter traveled
  return price
```

```python
taxi(0.560) # input is 560 meters = 0.560 km
```

# **Median of 3 values**

Write a function that takes 3 nos as parameters and returns the median value of hose parameters as its result.

Include a main program that reads 3 values from user and displays their median.

```python
# median of three numbers
def med(x1,x2,x3):
  mn=min(x1,x2,x3)
  mx=max(x1,x2,x3)
  md=(x1+x2+x3)-(mn+mx)
  return md
```

```python
x=med(99,-32,0)
print(x)
```

# **Reduce a Fraction to Lowest Terms**

Step 1: Find prime factors

6 = 2  3

63 = 3  3  7

Step 2: Find common prime factors

Common factor = 3

Step 3: Multiply common factors

i.e. GCD=3

Accept numerator and denomnator from user and divide both by GSD.

Whateveranswer you get as numerator and denominator is the lowest term.




image.png

```python
# fraction
def reduce_fraction(num,den):
  import math
  cd=math.gcd(num,den)
  num=num//cd
  den=den//cd
  return num,den
```

```python
z=reduce_fraction(6,63)
print(z)
```

```python
z=reduce_fraction(35,100)
print(z)
```

# **If Prime or not**

Check if number given as a parameter to a function is prime or not.

Prime number is divisible by 1 and itself only.

```python
def is_prime(n):
  if n<=1:
    return False
  for i in range(2,n):
    if n % i == 0:
      return False
  return True
```

```python
is_prime(9)
```

```python
is_prime(13)
```
