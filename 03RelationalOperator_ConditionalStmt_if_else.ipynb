# **Conditions using operators**

*   In programming, selection is the process of making decisions, allowing a program to choose between different execution paths
*   This concept is crucial for controlling flow and is usually represented using conditional statements like if, else in various programming languages.
*   Decision making naturally translates into boolean logic where True (1) represents condition is satisfied and False (0) represents condition is not satisfied



### **Conditional Statement**


*   take the input from user : input()
*   input() takes the value by default in the string format



### **if**
*   Check whether the number is positive



```python
a=int(input("Enter a value: "))
# check for positive number
if a>0:
   print("Positive Number")
```

```python
a=int(input("Enter a value: "))
# check for positive number
if a>0:
    print("Positive Number")

# check for negative number
if a<0:
    print("Negative Number")

# check for zero
if a==0:
    print("It is Zero")
```

### **if .... elif ... else**

*   if want to check multiple condition



```python
num=int(input("Enter either 10,20 or 30 "))
if num==10:
    print("TEN")
elif num==20:
    print("TWENTY")
elif num==30:
    print("THIRTY")
else:
    print("INVALID CHOICE")
```

```python
city=input("Enter a city either agra , delhi ,mumbai: ").lower()
print(city)
if city=="delhi":
    print("Red Fort")
elif city=="agra":
    print("Taj Mahal")
elif city=="mumbai":
    print("Gateway of India")
else:
    print("Wrong Choice")
```

### **nested if**

```python
num=int(input("Enter a value "))
if num>500:
  print("Entered Value is greater than 500")
  print("Check for the number is 600 , 650 , 700")
  if num==600:
    print("Number is 600")
  elif num==650:
    print("Number is 650")
  elif num==700:
    print("Number is 700")
  else: print("unknown")
else:
  print("Entered Value is less than 500")
```

```python
# Check if a number is divisible by 3 or 5 or by both or not by 3 and 5
num = int(input("Enter a number: "))

if num % 3 == 0 and num % 5 == 0:
    print(num, "is divisible by both 3 and 5")
elif num % 3 == 0:
    print(num, "is divisible only by 3")
elif num % 5 == 0:
    print(num, "is divisible only by 5")
else:
    print(num, "is not divisible by either 3 or 5")
```
