# **1. Left aligned Triangle Pattern**

1 * on 1st line, 2 * and 2nd line and so on

String repetition

```python
# Left aligned Triangle Pattern

n = 5
for i in range(1,n+1):
  print(i * "*") # i * '*' -> string repetition

```

```python
# same code using user input
n = int(input("Enter any positive number: "))
for i in range(1,n+1):
  print(i * "*")
```

```python
# same code using while loop

n = int(input("enter a value: "))
i = 1
while i<=n:
  print(i * "*")
  i = i + 1
```

# **6. Right aligned Triangle**

Repetition and concatenation

If you want to print 5 lines, on 1st line initially 4 spaces and 1 star will be printed.

On next line 3 spaces and 2 stars and so on.

So on 1st line (" " * 4)  + '*' * 1

On 2nd line (" " * 3)  + '*' * 2

On 3rd line (" " * 2)  + '*' * 3 and so on

```python
n = 5
for i in range(1,n+1):
  print(" " * (n-i) + '*' * i)
```

## **8. Magic Dates**

A magic date is a date where the day multiplied by the month = the last 2 digits of the year.

eg. June 10, 1960. Here June is 6th month. So 6*10=60. Last 2 digits of 1960 are 60.

So this is magic date.

How to get last 2 digits from 4 digits?

Divide 4 digits number by 100, you will get last 2 digist in ramainder.

Similarly 4 digit number if divided by 10, you will get last 1 digit in remainder, and so on

```python
# above code with diffent logic
def magic_date(d,m,y):
  y = str(y)
  y2 = y[2:4]
  if d * m == int(y2):
    print("It's a Magic Date")
  else:
    print("Not a Magic Date")
magic_date(8,9,1972) # try (10,7,1970), (10,6,1960),2/12/2024
```

```python
def magic_date(d,m,y):
  y = y % 100
  print(y)
  if d * m == y:
    print("It's a Magic Date")
  else:
    print("Not a Magic Date")
magic_date(8,9,1972)
```

# **9. Sorted Order**

WAP that reads integers form user and sotres them in a list.

Program should continue reading values until user enters 0.

Then it should display all values except 0 in ascending order.

Use either sort() or sorted() to sort the list.

```python
# Sorted - Ascending Order

x = int(input("Enter 1st number: "))
numbers = []
while x!=0:
  numbers.append(x)
  x = int(input("Enter next number: "))
print("\nUnsorted Numbers: ",numbers)
numbers.sort()
for i in numbers:
  print(i,end=" ")
```

# **10. Identify whether list is sorted in ascending order or not**

Write a function. If a list of 4 elements is entered in ascending order,

start comparing 1st two elements. If 1st element is greater than 2nd then return False

means list is not in ascending order.

Otherwise continue comparing 2nd and 3rd element and so on

```python
# List is Sorted in ascending order or Not
# input: ascending order list
def list_AscSorted(x):
    for i in range(len(x) - 1):
        if x[i] > x[i + 1]:
            return False
    return True
```

```python
list_AscSorted([1, 2, 5, 4])
```

```python
list_AscSorted([1, 6, 3, 4])
```

```python
# List is Sorted in descending order or Not
# input: descending order list
def list_DesSorted(x):
    for i in range(len(x) - 1):
        if x[i] < x[i + 1]:
            return False
    return True
```

```python
list_DesSorted([1, 2, 3, 4])
```

```python
list_DesSorted([6,5,4,2])
```
