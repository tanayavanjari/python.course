# **Inheritance**

- It is a process of creating new class from an alredy existing class.

*   Inheritance allows one class (child) to use the properties and methods of another class (parent).

- Existing class ---> Parent class or Main Class

- New Class ----> Child Class or Derived Class

**Syntax:**

**class Child(Parent):**

- **Types of Inheritance:**

    - **Single Inheritance :** child class is created from only one parent class

    - **Multiple Inheritance :** child class is created from multiple parent classes

    - **Multilevel Inheritance :** A class is derived from another derived class. Sequence of classes

    - **Hierarchical Inheritance :** multiple child classes are created from same parent class. All children share the same base class, but each child can have its own behavior.

    - **Hybrid Inheritance :** Its a combination of any two or more inheritance in a single program.

    It usually mixes:
*   Single
*   Multiple
*   Multilevel
*   Hierarchical





# **Single Inheritance:**

```python
# Single Inheritance:
# Parent class
class Person():
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display(self):
        print(f"Name: {self.name}, Age: {self.age}")
```

```python
x=Person("Amit",35)
```

```python
x.display()
```

```python
# Child class inheriting from Person
class Employee(Person):
    def __init__(self, name, age, salary): # this is constructor of Derived class

    # Call Constructor of Base/Super Class
    # Before setting parameters of Employee class we should call base class's init method.
    # Because Employee has all properties of Person(base) class i.e. 'name,age' and specific property of Employee as 'salary'.
    # To call init() method of Person class use super().__init__()

    # initialize super class and derived class attributes
        super().__init__(name, age) # it will initialize name and age of base class
        self.salary = salary # it will initialize type of derived class

    # define method of derived class
    def display_employee(self):
        self.display()  # Call parent class method
        print(f"Salary: {self.salary}")
        print(f"age: {self.age}")
        print(f"name: {self.name}")
```

```python
# Creating an Employee object
emp=Employee("Rajesh", 30, 50000)
```

```python
# Display Employee details
emp.display_employee()
```

```python
# invoking method of parent class
emp.display()
```

# **Multiple Inheritance:**

Child class is created from multiple parent classes

```python
# Parent class 1
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display_person(self):
        print(f"Name: {self.name}, Age: {self.age}")
```

```python
# Parent class 2
class Employee:
    def __init__(self, salary, position):
        self.salary = salary
        self.position = position

    def display_employee(self):
        print(f"Salary: {self.salary}, Position: {self.position}")
```

```python
# Child class inheriting from both Person and Employee
class Manager(Person, Employee):
    def __init__(self, name, age, salary, position, department):
        # Call constructors of both parents
        Person.__init__(self, name, age)
        Employee.__init__(self, salary, position)
        self.department = department

    def display_manager(self):
        self.display_person()  # Call method from 1st parent Person
        self.display_employee()  # Call method from 2nd parent Employee
        print(f"Department: {self.department}")
```

```python
mgr = Manager("Alice Johnson", 40, 80000, "Project Manager", "IT")
```

```python
mgr.display_manager()
```

```python
mgr.display_person()
```

```python
mgr.display_employee()
```

## **Parent Class**

```python
# Let us create Measurements Class
from math import pi
class measurements():
  # Adding object using init constructor
  def __init__(self,r,h,b):
    self.r=r
    self.h=h
    self.b=b
  # Method to compute area of circle
  def cir_area(self):
    carea=pi*(self.r**2)
    print("Area of Circle is",carea)
  # Method to compute area of triangle
  def tri_area(self):
    tarea=0.5*self.b*self.h
    print("Area of Triangle is",tarea)
  # Surface area of cylinder : TSA=2œÄr(r+h) = 2*œÄ*r*r + 2*œÄ*r*h
  def cyl_area(self):
    carea=2*pi*(self.r**2) + 2*pi*self.r*self.h
    print("Surface Area of Cylinder is",carea)
```

## **Child Class**

```python
class Volumes(measurements):
    # constructor of child class
    def __init__(self, r, h, b):
        super().__init__(r, h, b)   # call parent constructor

    # Volume of cylinder: œÄr¬≤h
    def cyl_volume(self):
        vol = pi * (self.r ** 2) * self.h
        print("Volume of Cylinder is", vol)

    # Volume of sphere: (4/3)œÄr cube
    def sphere_volume(self):
        vol = (4/3) * pi * self.r ** 3
        print("Volume of Sphere is", vol)
```

# **Multilevel Inheritance:**

A class is derived from another derived class.

```python
# Grandparent class
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display_person(self):
        print(f"Name: {self.name}, Age: {self.age}")
```

```python
# Parent class (Inheriting from Person)
class Employee(Person):
    def __init__(self, name, age, salary):
        super().__init__(name, age)  # Call parent constructor
        self.salary = salary

    def display_employee(self):
        self.display_person()  # Call method from Person class
        print(f"Salary: {self.salary}")
```

```python
# Child class (Inheriting from Employee)
class Manager(Employee):
    def __init__(self, name, age, salary, department):
        super().__init__(name, age, salary)  # Call Employee constructor
        self.department = department

    def display_manager(self):
        self.display_employee()  # Call method from Employee class
        print(f"Department: {self.department}")
```

```python
mgr = Manager("Ratan TATA", 40, 8000000, "IT")
```

```python
mgr.display_manager()
```

# **Hierarchical Inheritance:**

Multiple child classes are created from same parent class.

```python
# Parent Class
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display_person(self):
        print("Name:",self.name, "Age:", self.age)
```

# **Hybrid Inheritance:**

Hybrid inheritance is a combination of two or more types of inheritance, such as single, multiple, multilevel, or hierarchical inheritance in a single program. This allows for a flexible and structured way to manage relationships between classes.

```python
# Let us create Person Class (Parent Class)
class Person:
    def __init__(self, name, age, **kwargs): # **kwargs: keyword arguments
        self.name = name
        self.age = age

    def display_person(self):
        print(f"Name: {self.name}, Age: {self.age}")
```

```python
class Employee(Person):  # Single Inheritance
    def __init__(self, employee_id,**kwargs):
        super().__init__(**kwargs)
        self.employee_id = employee_id

    def display_employee(self):
        print(f"Employee ID: {self.employee_id}")

class Student(Person):  # Single Inheritance
    def __init__(self,student_id,**kwargs):
        super().__init__(**kwargs)
        self.student_id = student_id

    def display_student(self):
        print(f"Student ID: {self.student_id}")
# upto this Hierarchical Inheritance - One Parent and Multiple children

class Intern(Employee, Student):  # Multiple Inheritance (Hybrid) and Multilevel also
    def __init__(self, name, age, employee_id, student_id, duration):
        super().__init__(
            name=name,
            age=age,
            employee_id=employee_id,
            student_id=student_id
        )
        self.duration = duration

    def display_intern(self):
        self.display_person()
        self.display_employee()
        self.display_student()
        print(f"Internship Duration: {self.duration} months")
```

```python
# Example usage:
intern = Intern("Alice", 22, "E123", "S456", 6)
intern.display_intern()
```

# **Method Resolution Order (MRO):**

MRO (Method Resolution Order) is the order in which Python looks for a method in a class hierarchy when multiple classes are involved in inheritance

In multiple inheritance (1 child - multiple parents), every class in the chain must forward kwargs.

MRO for class Intern(Employee, Student):

Intern ‚Üí Employee ‚Üí Student ‚Üí Person ‚Üí object

Constructor flow:

Intern.__init__ ‚Üí Employee.__init__

Employee.__init__ ‚Üí Student.__init__

Student.__init__ ‚Üí Person.__init__

‚ùå STOP (chain breaks here)

üëâ object.__init__() is never reached

üëâ This violates cooperative multiple inheritance

```python
# Correct code
class Person:
    def __init__(self, name, age, **kwargs): # **kwargs: keyword arguments
        self.name = name
        self.age = age
        super().__init__(**kwargs)

    def display_person(self):
        print(f"Name: {self.name}, Age: {self.age}")

class Employee(Person):  # Single Inheritance
    def __init__(self, employee_id,**kwargs):
        super().__init__(**kwargs)
        self.employee_id = employee_id

    def display_employee(self):
        print(f"Employee ID: {self.employee_id}")

class Student(Person):  # Single Inheritance
    def __init__(self,student_id,**kwargs):
        super().__init__(**kwargs)
        self.student_id = student_id

    def display_student(self):
        print(f"Student ID: {self.student_id}")
# upto this Hierarchical Inheritance - One Parent and Multiple children

class Intern(Employee, Student):  # Multiple Inheritance (Hybrid) and Multilevel also
    def __init__(self, duration,**kwargs):
        super().__init__(**kwargs)
        self.duration = duration

    def display_intern(self):
        self.display_person()
        self.display_employee()
        self.display_student()
        print(f"Internship Duration: {self.duration} months")
```

```python
# Example usage:
intern = Intern(name="Alice",age=22,employee_id="E123",student_id="S456",duration=6)
intern.display_intern()
```

# **Protected Members (single underscore)**

Accessible in class and child classes

```python
# Parent Class
class Person:
    def __init__(self, name, age): # **kwargs: keyword arguments
        self._name = name
        self._age = age

    def display_person(self):
        print(f"Name: {self._name}, Age: {self._age}")

# Child Class
class Employee(Person):
  def __init__(self, name, age, salary):
        super().__init__(name, age)
        self.salary = salary
  def display_employee(self):
        print(f"Salary: {self.salary}")
        print(f"age: {self._age}")
        print(f"name: {self._name}")
```

# **Bank Account Example**

```python
class BankAccount:
    def __init__(self, account_number, balance):
        self.account_number = account_number  # Public attribute
        self.__balance = balance  # Private attribute (Encapsulation)
    # Getter method to access private attribute
    def get_balance(self):
        return self.__balance
    # Setter method to update private attribute
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited: {amount}. New Balance: {self.__balance}")
        else:
            print("Invalid deposit amount!")

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew: {amount}. Remaining Balance: {self.__balance}")
        else:
            print("Invalid withdrawal amount or insufficient balance!")

# Creating an object of BankAccount
acnt = BankAccount("123456789", 5000)

# Accessing public attribute
print(f"Account Number: {acnt.account_number}")

# Accessing private attribute using getter
print(f"Current Balance: {acnt.get_balance()}")

# Trying to modify private attribute directly (Not Recommended)
# account.__balance = 10000  # This won't actually modify the balance

# Using setter methods
acnt.deposit(10000)
acnt.withdraw(3000)

# Trying to access private attribute directly (This will fail)
# print(account.__balance)  # AttributeError: 'BankAccount' object has no attribute '__balance'
```

# **Polymorphism:**

Polymorphism allows the same method name to be used in different ways, depending on the object calling it.

i.e. One function behaves differently based on the object or parameter on which it works.

When we implement polymorphism using inheritance, it uses idea of Method OverridingThis increases flexibility and reusability.

```python
# Parent Class
class Person:
  # Define the constructor
  def __init__(self,name,age):
    self.name = name
    self.age = age

  # Define instance methods
  def say_hello(self):
    print(f'{self.name} says Hello from parent class.')
```

```python
# Child Class
class Astronaut(Person):
  # Constructor of the derived class
  def __init__(self,name,age,type):
    # Call the constructor of the base/super class
    super().__init__(name,age)
    # Initialise the derived class attributes
    self.type = type


  def launch_satellite(self):
    print(f'{self.name} is launching a satellite.')

  def say_hello(self):
    print(f'{self.name} the {self.type} says Hello from derived class.')
```

```python
p = Person('Mihir',25) # object of parent class
a = Astronaut('Raj',35,'Pilot') # object of child class
```

```python
p.say_hello()
```

```python
a.say_hello()
```

**say_hello() same method in base and derived class behaves differently depending on object we pass to it (name and type). This is polymorphism.**

# **Animal Example of Polymorphism**

```python
# Parent class - Hierarchical Inheritance - One parent, Multiple child classes
class Animal:
    def speak(self):
        return "Some sound"

# Child class 1
class Dog(Animal):
    def speak(self):
        return "Bark"

# Child class 2
class Cat(Animal):
    def speak(self):
        return "Meow"

# Function that demonstrates polymorphism
def animal_sound(animal):
    print(animal.speak())

# Creating objects
dog = Dog()
cat = Cat()

# Calling the same method on different objects
animal_sound(dog)
animal_sound(cat)
```
