8. Units of pressure. Read a pressure from the user in kilopascals. Program should return the equivalent pressure in pounds per square inch, millimeters of mercury and atmosphere. Use standandard conversion formulae for this.

a. kPa to pounds per square inch:
Pressure (psi) = Pressure (kPa) × 0.145038

OR Pressure (psi) = Pressure (kPa) ÷ 6.895 As 1 psi = 6.895 kPa

b. kPa to millimeters of Hg: mmHg = kPa × 7.50062

c. kPa to stmosphere: Pressure (atm) = Pressure (kPa) ÷ 101.325

```python
pre_kpa=float(input("Enter pressure in KiloPascals: "))

pre_psi=pre_kpa*0.145038
pre_mmhg=pre_kpa*7.50062
pre_atm=pre_kpa/101.325

print("Equivalent Pressure of {} kPa is {} pounds per square inch.".format(pre_kpa,pre_psi))
print("Equivalent Pressure of {} kPa is {} millimeters of mercury.".format(pre_kpa,pre_mmhg))
print("Equivalent Pressure of {} kPa is {} atmosphere.".format(pre_kpa,pre_atm))
```

9. Sum of digits in an integer. i.e. 1291 = 1+2+9+1 = 13

2nd Method: Keep on dividing number by 10 always and check remainder and quotients and take decision accordingly.

```python
n=int(input("Enter any number upto 4 digits:"))
# n = 1547
q1=n//1000 # Quotient after dividing by 1000
r1=n%1000 # remainder after dividing by 1000

q2=r1//100 # Quotient after dividing by 100
r2=r1%100 # remainder after dividing by 100

q3=r2//10 # Quotient after dividing by 10
r3=r2%10 # remainder after dividing by 10

print("Sum of digits of {} is {}".format(n,q1+q2+q3+r3))
```
