# **__init__**

In Python, the __init__ method is a special method that is automatically called when a new
object is created from a class.


Its primary role is to initialize the object — in other words, set up the object with initial data


(also called attributes or instance variables).

You can think of it as a constructor in other programming languages.

# **Define a Class**

```python
# Let us create a person class
class person():
  # Let us add objects using __init__ constructor.
  def __init__(self,name,age,height,nation): # in bracket we have objects
    self.name=name
    self.age=age
    self.height=height
    self.nation=nation
  # Let us create a method to print name of person
  def pname(self):
    print(self.name)
  def page(self):
    print(self.age)
  def pheight(self):
    print(self.height)
  def pnation(self):
    print(self.nation)
  def pdetails(self):
    print("name:",self.name)
    print("age:",self.age)
    print("Height",self.height)
    print("Nationality:",self.nation)
```

# **Create Objects based on the Class**

```python
# Create instances of class
x=person("ravi",26,157,"Indian")
y=person("Hema",24,151,"Indian")
```

# **Call the method of class for an instance**

```python
# let us call the method for an instance
y.pdetails()
```

```python
x.pdetails()
```

```python
x.pname()
```

## **Define a class Person.**

Attributes: name, age

Methods: say_hello() -> displays 'name says hello'.

is_eligible() -> displays 'Eligible' if age>=18, else 'Underage'

```python
# Define a class
class Person:
  # Define the constructor
  def __init__(self,name,age): # Constructor of class
    # create and initialize the object
    self.name = name
    self.age = age
  # Define instance method
  def say_hello(self):
    print(f'{self.name} says hello')

  # let us define another method
  def is_eligible(self):
    if self.age >= 18:
      return 'Eligible'
    else:
      return 'Underage'
```

```python
# create an object of type Person
p1 = Person('Tom',23)
```

```python
type(p1) # it is an object of type Person
```

```python
# call method
p1.say_hello()
```

```python
p1.is_eligible()
```

```python
p2 = Person('Mary',17)
```

```python
p2.say_hello()
```

```python
p2.is_eligible()
```

```python
p1.name
```

## **Define a class called Student**

Attributes: name, marks, course

Method: grades() ->

if marks >= 80 Distinction

marks >= 60 First Class

marks >= 40 Second Class

else Fail

```python
# Class name start with capital letter... it's a naming convention
class Student:
  def __init__(self,name,marks,course): #self is always 1st parameter for the constructor
  # create and initialize the object
    self.name = name
    self.marks = marks
    self.course = course

  # define a method
  def grades(self):
    if self.marks >= 80:
      return 'Distinction'
    elif self.marks >= 60:
      return 'First Class'
    elif self.marks >= 40:
      return 'Second Class'
    else:
      return 'Fail'
```

```python
# Create students of type Student
s1 = Student('Tom',75,'Python')
```

```python
s1.__dict__ # see properties of 1st student
```

```python
s2 = Student('Mary',45,'Data Science')
```

```python
s2.__dict__ # see properties of 2nd student
```

## **Define a class calc**

Attributes: x,y

4 Methods: addition() displays 'Addition:' result, subtract(), mul(), div()

```python
class calc():
  def __init__(self,x,y):
    self.x=x
    self.y=y
  def addition(self):
    print("sum:",self.x+self.y)
  def subtract(self):
    print("Difference:",self.x-self.y)
  def prod(self):
    print("Product:",self.x*self.y)
  def div(self):
    print("Division:",self.x/self.y)
```

```python
p=calc(10,20)
```

```python
p.addition()
```

## **Define a class point to calculate distance between two points**

Attributes: x1,y1,x2,y2

Methods: distance() using formula: distance=sqrt[(x1-x2)square + (y1-y2)square]

```python
# distance between two points
import math
class point():
  def __init__(self,x1,y1,x2,y2):
    self.x1=x1
    self.y1=y1
    self.x2=x2
    self.y2=y2
  def distance(self):
    d=math.sqrt((self.x1-self.x2)**2+(self.y1-self.y2)**2)
    print(d)
```

```python
p=point(1,2,3,4)
```

```python
p.distance()
```

# **Creating Multiple Independent Classes:**


```python
# let us create first class
class one():
  def __init__(self,num1,num2):
    self.num1=num1
    self.num2=num2
  def numbers(self):
    print(self.num1,self.num2)

# Let us create another class
class two():
  def __init__(self,name1,name2):
    self.name1=name1
    self.name2=name2
  def names(self):
    print(self.name1,self.name2)
```

```python
obj1=one(2,4)
obj2=two('A','B')
```

```python
obj1.numbers()
```

```python
obj2.names()
```
