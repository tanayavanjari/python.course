# **Tuple**
*   A tuple is a built-in data structure in Python used to store multiple values in a single variable
*   A tuple is an ordered, immutable collection of elements.



# **Key Properties of Tuple:**

✅ Tuples are created using round brackets ()

✅ Ordered (index-based)

❌ Immutable (cannot be changed)

✅ Allows duplicate values

✅ Can store different data types

✅ Faster than lists

✅ Uses less memory than lists


```python
x=tuple()
print(x)
print(type(x))
```

```python
y=tuple("snehal")
print(y)
print(type(y))
```
